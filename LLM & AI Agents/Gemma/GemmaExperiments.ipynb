Summarize the novel text provided by the user using the Gemma model, splitting the text into chunks if necessary.

## Load the text

Get the text of the novel into a string format.


**Reasoning**:
Assign the provided novel text to a multiline string variable.



```python
novel_text = """
It was a dark and stormy night. The wind howled around the ancient castle, and rain lashed against the stone walls. Inside, Lady Eleanor shivered by the flickering fire, her thoughts as gloomy as the weather. She had received a cryptic message earlier that day, a message that spoke of a forgotten prophecy and a hidden treasure. A treasure that, if found, could change the fate of her family and the entire kingdom.

The message had been delivered by a hooded figure who vanished as quickly as he appeared. It was written on parchment, in a hand she didn't recognize. It spoke of trials and guardians, of a journey to the heart of the Whispering Woods and the peak of the Silent Mountains. Eleanor, though slight in build, possessed a spirit as strong as the castle stones. She knew she had to embark on this quest. Her family's honor, and perhaps their lives, depended on it.

The next morning, as the storm subsided and the first rays of dawn pierced the clouds, Eleanor prepared for her journey. She donned practical clothes, a sturdy cloak, and packed a small bag with essentials. Her loyal wolfhound, Shadow, sensed her mistress's resolve and stood by her side, a silent guardian. Leaving a note for her trusted steward, she slipped out of a secret passage, the castle gates too public for her clandestine departure.

The Whispering Woods lived up to their name. The trees were ancient and tall, their leaves rustling secrets in the breeze. Strange eyes watched from the shadows, and the path was winding and uncertain. Eleanor relied on Shadow's keen senses and her own intuition. She encountered puzzles left by the prophecy's creators – riddles etched into stones, pathways that shifted and changed. With each solved riddle, a part of the path became clearer, and her determination grew stronger. Shadow, with an occasional low growl, warned her of unseen dangers.

After days of travel, they emerged from the woods and saw the Silent Mountains in the distance, their peaks shrouded in mist. The transition was stark; the air grew thin and cold, and the terrain became steep and treacherous. The prophecy spoke of a 'Guardian of Stone' on the mountainside. Eleanor knew this would be her greatest challenge yet.

As they ascended, the wind whipped around them, and the silence was broken only by the crunch of their boots on the rocky path and Shadow's occasional panting. They found the Guardian, a massive stone golem, its eyes glowing with an internal light. It blocked the narrow path to the peak. The prophecy stated, "To pass the stone, speak its true name." Eleanor racked her brain, remembering ancient texts and forgotten lore. Shadow stood ready, a low growl rumbling in his chest. Eleanor looked into the golem's eyes and a name, a name whispered on the wind, came to her. She spoke it aloud, a single word echoing in the mountain air. The golem's light faded, and it stepped aside, revealing the path.

At the peak, a small, unassuming stone altar stood, bathed in the light of the rising sun. On the altar lay a single, intricately carved wooden box. This was the treasure. As Eleanor reached for it, the same hooded figure who delivered the message appeared. He lowered his hood, revealing a kind face, lined with age and wisdom. "You have proven yourself worthy, Lady Eleanor," he said. "The prophecy is fulfilled. This box contains not gold or jewels, but knowledge – the knowledge of your family's true lineage and the ancient power that flows through your veins. Use it wisely."

With the wooden box in hand, Eleanor and Shadow began their descent, the weight of newfound knowledge settling upon her. The journey back was less perilous; the woods seemed less threatening, the mountains less daunting. She returned to her castle, not as the shivering lady who left, but as a woman who understood her past and was ready to face the future. The castle and the kingdom would never be the same.
"""
print(f"Novel text loaded. First 100 characters:\n{novel_text[:100]}...")
```

## Split the text

### Subtask:
Divide the novel text into smaller chunks that fit within the model's context window.


**Reasoning**:
To process the novel text with the Gemma model, it needs to be split into smaller chunks. This code will split the text into chunks of approximately 500 tokens, respecting sentence boundaries.



```python
import re

def split_text_into_chunks(text, chunk_size):
    """Splits text into chunks of approximately chunk_size tokens."""
    sentences = re.split(r'(?<=[.!?])\s+', text)
    chunks = []
    current_chunk = ""
    for sentence in sentences:
        if len((current_chunk + " " + sentence).split()) <= chunk_size:
            if current_chunk:
                current_chunk += " " + sentence
            else:
                current_chunk = sentence
        else:
            chunks.append(current_chunk)
            current_chunk = sentence
    if current_chunk:
        chunks.append(current_chunk)
    return chunks

# Assuming a chunk size of 500 tokens as a reasonable estimate for Gemma 2B
chunk_size = 500
text_chunks = split_text_into_chunks(novel_text, chunk_size)

print(f"Generated {len(text_chunks)} chunks.")
print(f"First chunk starts with:\n{text_chunks[0][:200]}...")
```

## Summarize chunks

### Subtask:
Iterate through the chunks and use the Gemma model to generate a summary for each chunk.


**Reasoning**:
Initialize an empty list to store summaries and then iterate through the text chunks, generate a summary for each using the agent, and append it to the list.



```python
summaries = []
for chunk in text_chunks:
    prompt = f"Summarize the following text: {chunk}"
    response = agent(prompt, max_length=50, max_new_tokens=50) # Explicitly set max_new_tokens
    summaries.append(response[0]['generated_text'])

print(f"Generated {len(summaries)} summaries.")
```

## Display summaries

### Subtask:
Show the generated summaries to the user.


**Reasoning**:
Iterate through the generated summaries and print each one with a label indicating the chunk number.



```python
for i, summary in enumerate(summaries):
    print(f"Summary for Chunk {i + 1}:")
    print(summary)
    print("-" * 20)
```
