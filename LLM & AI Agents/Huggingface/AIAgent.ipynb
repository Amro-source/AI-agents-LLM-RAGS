```python
from transformers import pipeline

agent = pipeline("text-generation", model="HuggingFaceH4/zephyr-7b-beta")
response = agent("Explain AI agents simply", max_length=100)
print(response[0]['generated_text'])
```

```python
response = agent("Explain Mediterrean diet to me simply", max_length=100)
print(response[0]['generated_text'])
```

```python
from transformers import pipeline

# Use the smaller gemma-2b-it model
agent = pipeline("text-generation", model="google/gemma-2b-it")

# You can replace the prompt with your desired text generation task
prompt = "Explain AI agents simply"
response = agent(prompt, max_length=100)
print(response[0]['generated_text'])
```
